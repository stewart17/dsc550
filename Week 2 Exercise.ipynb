```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Introduction- Student Performance Dataset

The Student Performance dataset tracks students' scores on math, reading, and writing. It also contains features of the student such as their gender, race, parental level of education, etc. 

```python
# import student performance dataset
performance = pd.read_csv("StudentsPerformance.csv")
performance.head()
```

## Questions

1. Does a student's parental level of education influence the student's test scores?

## Graph 1-  Distribution of Scores

```python
# histogram of math scores
plt.hist(performance["math score"], label="Math")
# histogram of reading scores
plt.hist(performance["reading score"], label="Reading")
# histogram of writing scores
plt.hist(performance["writing score"], label = "Writing")
plt.xlabel("Test Score")
plt.ylabel("Frequency")
plt.title("Distribution of Test Scores (Math, Reading , Writing)")
plt.legend(loc="upper left")
plt.show()
```

This first graph shows the distribution of scores between math reading and writing. As can be seen, all three tests almost follow a normal distribution, with a slight negative skew. There appear to be a few outliers on the lower end of the distribution.

## Graph 2- Distribution of Parental Level of Education

```python
# histogram of parental level of education
plt.hist(performance["parental level of education"], label= "Parental Level of Education")
plt.xticks(rotation=45)
plt.ylabel("Frequency")
plt.title("Distribution of Parental Level of Education")
plt.show()
```

The above graph shows the frequency of students whose parents achieved a certain level of education. There is wide variation in what a parents level of education achieved. If we are to draw a substantive claim about the affect a parents' education has on a student's test performance, we should be careful not to look at the absolute number of high scores for each level, since by default, the categories with more frequency may have more scores by sheer numbers. We must look at the proportion of high scores compared to the amount in that category

## Graph 3- Parental Level of Education vs Scores

```python
# box plot of math scores
performance.boxplot(column = "math score", by = "parental level of education", figsize=(10, 7))
plt.xticks(rotation=45)
plt.title("Math Score Grouped by Parental Level of Education")
plt.ylabel("Math Score")
# box plot of reading scores
performance.boxplot(column = "reading score", by = "parental level of education", figsize=(10, 7))
plt.ylabel("Reading Score")
plt.title("Reading Score Grouped by Parental Level of Education")
# box plot of writing scores
performance.boxplot(column = "writing score", by = "parental level of education", figsize=(10, 7))
plt.ylabel("Writing Score")
plt.title("Writing Score Grouped by Parental Level of Education")
plt.show()
```

This box plot shows test scores grouped by parental level of education. It shows that across the board, students with a parental level of education of master's degrees had a higher median score for all three tests. In fact, for the most part, the median scores are highest in descending order of highest level of education (master's, bachelor's, etc). This suggests that parental level of education does have some influence on writing scores. 

## Conclusion

The above visualizations help provide insight into what factors influence a student's test score. At least from the above, we can come to the preliminary conclusion that the parental level of education has a positive relationship with a student's test scores. Of course, this does not mean that the parental level of education is the cause of a student's test score. The level of education could just be a proxy for a third explanatory factor. 

Further visualization would investigate other factors and compare them to one another to see which has the highest explanatory power.
