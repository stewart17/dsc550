```python
import pandas as pd
```

```python
# load data frame
game_sales = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv")
```

```python
# display first ten rows of data
game_sales.head(10)
```

```python
# number of rows & columns
game_sales.shape
```

Each row is a video game and each column provides particular info of that video game

```python
# top 5 games by critic score
game_sales.sort_values(by="Critic_Score", ascending=False).head(5)
```

```python
# number of video games in each genre
game_sales.groupby("Genre")["Name"].count()
```

```python
# first five games on SNES
game_sales[game_sales["Platform"]=="SNES"].sort_values(by="Year_of_Release").head(5)
```

```python
# find sum global sales for each pubisher
top_sales = game_sales.groupby("Publisher")["Global_Sales"].sum()
# create df of top sales
top_sales_df = pd.DataFrame(top_sales)
# show top 5 publishers with highest total global sales
top_sales_df.sort_values(by="Global_Sales", ascending=False).head(5)
```

```python
# new column: percentage of global sales from NA
game_sales["Pct_NA_Sales"] = game_sales["NA_Sales"] / game_sales["Global_Sales"] * 100
# display first five rows of data frame
game_sales.head()
```

```python
# number NaN entries in each column
game_sales.isna().sum()
```

```python
# trying to calculate median score of all video games
game_sales["User_Score"].median()
```

```python
# replacing strings with NaN
game_sales["User_Score"] = pd.to_numeric(game_sales["User_Score"], errors ="coerce")
# calculating median score
median_score = game_sales["User_Score"].median()
median_score
```

```python
# replacing nan's w/ median
game_sales["User_Score"] = game_sales["User_Score"].fillna(median_score)
game_sales
```
